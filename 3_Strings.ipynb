### String indexing 

```python
#stringlerde ilk karakterin indisi 0(sfr)'dr. Stringler dizi(liste)
#gibi her karakterine ulalabilir.
word = 'Python'
print(word[0]) #ilk karakter
print(word[5]) #son karakter
print(word[0][0][0][0])
```

```python
#son elemandan baland zaman indeksler negatif saylarla ifade edilir.
print(word[-1])  #son karakter
print(word[-2])  #sondan ikinci karakter
print(word[-6])  #ilk karakter
print(word[-7])
```

```python
#metin iinden bir blm de alnabilir.
print(word)
print(type(word[0:0]))
print(word[0:2])  # 0.indisten 2.indise kadar(2. indeks dahil deildir)
print(word[2:5])  # 2.indisten 5.indise kadar(2. indeks dahil deildir)
```

```python
print(word[:])  #Tm elemanlar listelenir
print(word[2:]) #2.indisinden balayarak listenin alan elemanlar listelenir
print(word[:2]) #listenin bandan balayarak 2.indisli elemana kadar.
                #2.indisli eleman dahil deildir.
```
