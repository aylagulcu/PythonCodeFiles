### Mathematical Operations: 

    / division (always returns a float) 
    //  integer divivision
    %  remainder
    **  calculates powers
    =  assignment

```python
>>> 17 / 3  # classic division returns a float
```

```python
>>> 17 // 4  # floor division discards the fractional part, return type depends
```

```python
>>> 17 % 3  # the % operator returns the remainder of the division
```

```python
>>> 5 * 3 + 2 # operator precedence
```

```python
>>> 2 ** 7  # 2 to the power of 7
```

```python
>>> width = 20
>>> height = 5 * 9
>>> area= width * height
print(area)
height = 20
print(area)
area= width * height
print(area)
```
